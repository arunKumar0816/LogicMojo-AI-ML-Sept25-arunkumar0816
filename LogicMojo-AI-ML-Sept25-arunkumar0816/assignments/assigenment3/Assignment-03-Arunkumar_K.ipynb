#### 6. What are the five most common County names in the U.S.?

```python

unique_geography = pop[['County', 'State']].dropna(subset=['County']).drop_duplicates()
top_five_counties = unique_geography['County'].str.replace(' County', '', regex=False).value_counts().head(5)
print(top_five_counties)
```

#### 7. What are the top 5 most populated Counties according to the 2010 Census?

```python
# Write your code here
top_5_2010 = pop[pop['Year'] == 2010].dropna(subset=['County']).sort_values(by='Population', ascending=False).head(5)
print(top_5_2010[['County', 'Population']])
```

#### 8. What are the top 5 most populated States according to the 2010 Census?

```python
top_5_2010 = pop[pop['Year'] == 2010].dropna(subset=['State']).sort_values(by='Population', ascending=False).head(5)
print(top_5_2010[['State', 'Population']])

```

#### 9. How many Counties have 2010 populations greater than 1 million?

```python
counties_2010 = pop[(pop['Year'] == 2010) & (pop['County'].notna())]
count = len(counties_2010[counties_2010['Population'] > 1000000])

print(f"Number of counties with 2010 population > 1,000,000: {count}")

```

#### 10. How many Counties don't have the word "County" in their name?

```python
unique_counties = pop['County'].dropna().unique()
no_county_word = [name for name in unique_counties if 'County' not in name]

print(f"Count: {len(no_county_word)}")


```

#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
counties_change = pop[pop['County'].notna()][['County', 'Percent_Change_2010_2017']].drop_duplicates()
print(counties_change.head())

```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
state_totals = pop[pop['State'].notna() & pop['County'].isna()]
p10 = state_totals[state_totals['Year'] == 2010][['State', 'Population']].rename(columns={'Population': 'Pop2010'})
p17 = state_totals[state_totals['Year'] == 2017][['State', 'Population']].rename(columns={'Population': 'Pop2017'})
state_growth = pd.merge(p10, p17, on='State')
state_growth['Percent_Change'] = ((state_growth['Pop2017'] - state_growth['Pop2010']) / state_growth['Pop2010']) * 100
top_states = state_growth.sort_values(by='Percent_Change', ascending=False)

print(top_states.head())

```
