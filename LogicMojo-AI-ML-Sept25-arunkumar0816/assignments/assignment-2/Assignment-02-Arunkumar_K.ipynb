```python
def cyclic_pattern(n):
     
    # pattern = [1,2,3]
    # cyclic_copy = pattern*3
    # return cyclic_copy

    pattern = np.array([1,2,3],dtype=int)
    cyclic_copy= np.tile(pattern,n)
    return cyclic_copy

n= int(input("Enter the value of n: "))
print(cyclic_pattern(n))
```

## Problem: 6

Create an array $x$ with 100 points in the interval $[-1,1]$ (the points must be uniformly spaced). Then create another array $y$. 

Where $y = x^2 sin(x^2) + x$. 

Try to plot $y$ vs $x$ using matplotlib.

```python
arr= np.linspace(1,-1,100)
arr2 = (arr**2) * np.sin(arr**2) + arr

plt.style.use('seaborn-v0_8-whitegrid')
plt.Figure(figsize=(10,6))
plt.plot(arr,arr2,label=r'$row2 = row^2\sin(row^2) + row$', color='purple')
plt.title(r'Plot of $row2 = row^2\sin(row^2) + row$')
plt.xlabel(r'$x$ (100 uniformly spaced points in $[-1, 1]$)')
plt.ylabel(r'$y$')
plt.legend()
plt.grid(True)
plt.savefig('plot_x_y_function.png')

```

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
def compute_array_sums(rows, columns,seed=None):

    if seed is not None:
        np.random.seed(seed)

    arr= np.random.rand(rows,columns)
    sum_of_entries = np.sum(arr)
    sum_of_rows = np.sum(arr,axis=0)
    sum_of_columns = np.sum(arr,axis=1)

    return{

        'array': arr,
        'sum_of_entries': sum_of_entries,
        'sum_of_rows': sum_of_rows,
        'sum_of_columns': sum_of_columns
    }


compute_array_sums(3,5)
```
